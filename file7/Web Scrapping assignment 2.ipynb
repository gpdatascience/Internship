```python
Question-1. Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note: All of the above steps have to be done in code. No step is to be done manually
```

# ANS-1

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize Chrome WebDriver
chrome_service = ChromeService("path_to_chromedriver")
chrome_options = ChromeOptions()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# Step 1: Get the webpage
driver.get("https://www.shine.com/")

# Step 2: Enter job title and location
job_title_input = driver.find_element(By.ID, "search-box-keyword")
job_title_input.clear()
job_title_input.send_keys("Data Scientist")

location_input = driver.find_element(By.ID, "search-box-location")
location_input.clear()
location_input.send_keys("Bangalore")

# Step 3: Click the search button
search_button = driver.find_element(By.ID, "btn-search")
search_button.click()

# Step 4: Scrape data for the first 10 jobs results
job_data = []

# Wait for the search results to load
wait = WebDriverWait(driver, 10)
job_results = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "search-card")))

for job_result in job_results[:10]:
    job_title = job_result.find_element(By.CLASS_NAME, "job-title").text
    job_location = job_result.find_element(By.CLASS_NAME, "job-location").text
    company_name = job_result.find_element(By.CLASS_NAME, "company-name").text

    job_data.append({
        "Job Title": job_title,
        "Job Location": job_location,
        "Company Name": company_name
    })

# Step 5: Create a dataframe of the scraped data
df = pd.DataFrame(job_data)

# Display the dataframe
print(df)

# Close the WebDriver
driver.quit()
```

# Answer-3

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open Shine.com
driver.get("https://www.shine.com/")

# Find the search field and enter data
job_title_input = driver.find_element(By.CLASS_NAME,"searchBarInput ")
job_title_input.send_keys("Data Analyst")

# Click the search button
search_button = driver.find_element(By.CLASS_NAME,"search-btn")
search_button.click()

# Apply location filter (Delhi/NCR)
location_filter = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//label[@for="locDelhi"]'))
)
location_filter.click()

# Apply salary filter (3-6 lakhs)
salary_filter = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//label[@for="salary-3-6-lakhs"]'))
)
salary_filter.click()

# Wait for the results to load
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, "wht-shd-bg"))
)

# Scrape data for the first 10 jobs
job_data = []
job_results = driver.find_elements_by_class_name("wht-shd-bg")

for job in job_results[:10]:
    job_title = job.find_element_by_class_name("wht-dtl").text
    job_location = job.find_element_by_class_name("loc").text
    company_name = job.find_element_by_class_name("snp").text
    experience_required = job.find_element_by_class_name("exp").text
    job_data.append([job_title, job_location, company_name, experience_required])

# Create a DataFrame
df = pd.DataFrame(job_data, columns=["Job Title", "Job Location", "Company Name", "Experience Required"])

# Print the DataFrame
print(df)

# Close the web browser
driver.quit()
```

# Answer-6

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import warnings
warnings.filterwarnings('ignore')

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Define the URL for searching sneakers on Flipkart
url = "https://www.flipkart.com/"

# Open the URL in the web browser
driver.get(url)

# Close the login popup if it appears
try:
    close_popup = driver.find_element(By.XPATH, "//button[@class='_2KpZ6l _2doB4z']")
    close_popup.click()
except:
    pass

# Find the search field and enter "sneakers"
search_field = driver.find_element(By.NAME, "q")
search_field.send_keys("sneakers")
search_field.submit()

# Initialize lists to store scraped data
brands = []
descriptions = []
prices = []

# Scrape data from the first 100 sneaker listings
while len(brands) < 100:
    # Find all the sneaker listings on the current page
    listings = driver.find_elements(By.CLASS_NAME, "_1AtVbE")

    for listing in listings:
        try:
            # Extract brand, description, and price
            brand = listing.find_element(By.CLASS_NAME, "_2WkVRV").text
            description = listing.find_element(By.CLASS_NAME, "IRpwTa").text
            price = listing.find_element(By.CLASS_NAME, "_30jeq3").text

            # Append data to the respective lists
            brands.append(brand)
            descriptions.append(description)
            prices.append(price)
        except Exception as e:
            print(f"An error occurred: {str(e)}")

    # Find and click the link to the next page
    try:
        next_button = driver.find_element(By.CLASS_NAME, "_1LKTO3")
        if next_button.is_enabled():
            next_button.click()
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "_1AtVbE")))
        else:
            break
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Create a DataFrame from the scraped data
data = {'Brand': brands, 'Product Description': descriptions, 'Price': prices}
df = pd.DataFrame(data)
df.head(100)
#driver.quit()
```

# Answer-7

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open Amazon.in
driver.get("https://www.amazon.in/")

# Find the search field and enter "Laptop"
search_field = driver.find_element(By.ID, "twotabsearchtextbox")
search_field.send_keys("Laptop")
search_field.send_keys(Keys.RETURN)

# Wait for search results to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "s-result-item")))

# Find and click the CPU Type filter for "Intel Core i7"
cpu_filter = driver.find_element(By.XPATH, "//li[@aria-label='Intel Core i7']")
cpu_filter.click()

# Wait for the filter to be applied
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "s-result-item")))

# Scrape data for the first 10 laptops
laptop_data = []

laptops = driver.find_elements(By.CSS_SELECTOR, ".s-result-item")
for laptop in laptops[:15]:
    try:
        title = laptop.find_element(By.CSS_SELECTOR, "h2 a span").text
        ratings = laptop.find_element(By.CSS_SELECTOR, ".a-icon-star-small .a-icon-alt").get_attribute("innerHTML")
        price = laptop.find_element(By.CSS_SELECTOR, ".a-price-whole").text

        laptop_data.append([title, ratings, price])
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Create a DataFrame from the scraped data
df = pd.DataFrame(laptop_data, columns=["Title", "Ratings", "Price"])

# Print the DataFrame
df

# Close the web browser

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Initialize the WebDriver (you need to specify the path to your webdriver executable)
# Make sure to download the appropriate webdriver for your browser (e.g., ChromeDriver)
driver = webdriver.Chrome()

# Step 1: Get the webpage
url = "https://www.azquotes.com/"
driver.get(url)

# Step 2: Click on "Top Quotes"
top_quotes_link = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
top_quotes_link.click()

# Wait for some time to load the page
time.sleep(5)

# Initialize an empty list to store the quotes data
quotes_data = []

# Step 3: Scrape the data (Quote, Author, Type Of Quotes)
while len(quotes_data) < 1000:
    quote_elements = driver.find_elements(By.XPATH, "//div[@class='wrap-block']")
    
    for element in quote_elements:
        quote_text = element.find_element(By.CLASS_NAME, "title").text
        author_name = element.find_element(By.CLASS_NAME, "author").text
        type_of_quotes = element.find_element(By.CLASS_NAME, "tags").text

        quotes_data.append({
            "Quote": quote_text,
            "Author": author_name,
            "Type Of Quotes": type_of_quotes
        })

    # Scroll down to load more quotes
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Wait for some time to load more quotes
    time.sleep(3)

# Close the WebDriver
#driver.quit()

# Print the first 10 quotes as a sample
df=pd.DataFrame(quotes_data)
df

# You can save the data to a file or process it further as needed
```
